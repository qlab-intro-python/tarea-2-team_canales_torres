## Summarizing grouped data

14. Calculate average delays by destination for flights originating in NYC. Create a variable which ranks destinations from worst to best, where 1 is the worst rank.

```python
# Solution
#Veamos los vuelos que parten de los aeropuertos JFK, EWR y LGA.
datos_nyc = df[df['origin'].isin(['JFK', 'EWR', 'LGA'])]
datos_nyc

#Retrasos promedio:
retrasos_promedio = (datos_nyc.groupby('dest')['dep_delay'].mean() + datos_nyc.groupby('dest')['arr_delay'].mean()) / 2

#Aqui vemos la rpta: Se creo una variable que ordena los destinos de peor a mejor segun su retraso promedio
retrasos_promedio_ordenados = retrasos_promedio.sort_values(ascending=False)
top_10_retrasos = retrasos_promedio_ordenados.nlargest(10)
ranking = top_10_retrasos.rank(method='max', ascending=False)
datos_rankeados = pd.DataFrame({'Retraso Promedio': top_10_retrasos, 'Ranking': ranking})
datos_rankeados
```

15. Which plane (`tailnum`) has the most minutes of delays total?

```python
# Solution
# Calculamos el promedio de minutos de retraso tanto en la salida como en la llegada para cada avión
promedio_retraso_aviones = (datos_nyc.groupby('tailnum')['dep_delay'].mean() + datos_nyc.groupby('tailnum')['arr_delay'].mean()) / 2

# Ordenamos los resultados de manera descendente para identificar el avión con el mayor promedio de retraso
promedio_retraso_aviones = promedio_retraso_aviones.sort_values(ascending=False)

# Mostramos el avión con el mayor promedio de minutos de retraso
promedio_retraso_aviones.head(1)

```

20. Write a code which calculates your delay measures separately by airline. Do particular airlines perform poorly on multiple metrics?

```python
# Solution
#Calculamos las medidas de retraso individualmente para cada aerolínea
promedio_retraso_salida = df.groupby('carrier')['dep_delay'].mean()
promedio_retraso_llegada = df.groupby('carrier')['arr_delay'].mean()

# Construir un DataFrame con los resultados
df_medidas_retraso = pd.DataFrame({
    'Aerolínea': promedio_retraso_salida.index,
    'Promedio Retraso en Salida': promedio_retraso_salida.values,
    'Promedio Retraso en Llegada': promedio_retraso_llegada.values
})

# Ordenar el DataFrame por ambas métricas 
df_medidas_retraso = df_medidas_retraso.sort_values(by=['Promedio Retraso en Salida', 'Promedio Retraso en Llegada'], ascending=False)

# Imprimir el resultado para análisis
print(df_medidas_retraso)

```
